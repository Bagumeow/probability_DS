## 1. BÀI TẬP VỀ BIẾN NGẪU NHIÊN LIÊN TỤC

Xét hàm mật độ xác suất $f(x)$ của biến ngẫu nhiên liên tục $X=x$. Ta đã biết 
$$\int_{-\infty }^{+\infty }{f(x)dx}=1.$$ Để tính xác suất $P(a\le X\le b),$ ta tính $\int_{a}^{b}{f(x)dx}$.
Ta cũng có:
1) Hàm phân bố xác suất: $F(x)=P(X<x)=\int_{-\infty }^{x}{f(t)dt}$ (bản chất là nguyên hàm của $f(x)$, tức là $f(x)={F}'(x)$). Khi đó, giới hạn tại $-\infty ,+\infty $ của $F(x)$ lần lượt là $0,1.$ <br>
2) Kỳ vọng: $E(X)=\int_{-\infty }^{+\infty }{xf(x)dx}$; <br>
3) Phương sai: $D(X)=\int_{-\infty }^{+\infty }{{{(x-E(X))}^{2}f(x)}dx}=\int_{-\infty }^{+\infty }{{{x}^{2}}f(x)dx}-{{(E(X))}^{2}}$;<br>
4) Độ lệch chuẩn: $\sigma (X)=\sqrt{D(X)}$. <br>
5) Mốt (mode): giá trị ${{x}_{0}}$ sao cho $f(x)$ đạt max tại đó. <br>
6) Trung vị (median): giá trị ${{x}_{0}}$ sao cho $F({{x}_{0}})=\frac{1}{2}$ hay $\int_{-\infty }^{{{x}_{0}}}{f(x)dx}=\frac{1}{2}.$

## Phân bố đồng thời và hệ số tương quan.
Phân bố xác suất đồng thời cho hai đại lượng ngẫu nhiên rời rạc $X, Y$ sẽ được ký hiệu bởi $p_{ij} = P(X=x_i, Y = y_i)$. Do tính rời rạc của các biến, ta có thể liệt kê chúng thành bảng có dạng sau:

| $X$ | $y_1$ | $y_2$ | $\ldots$ | $y_n$
| --- | --- | --- | --- | ---
| $x_1$ | $p_{11}$ | $p_{12}$ | $\ldots$ | $p_{1n}$
| $x_2$ | $p_{21}$ | $p_{22}$ | $\ldots$ | $p_{2n}$
| ... | ... | ... | ... | ...
| $x_m$ | $p_{m1}$ | $p_{m2}$ | $\ldots$ | $p_{mn}$

Ta biết rằng: <br>
1. Tổng các số trên bảng vừa bằng $1$. <br>
2. Để tính $P(X=x_i)$, ta tính tổng các số ở hàng $i$.
3. Để tính $P(Y=y_i)$, ta tính tổng các số ở cột $i$.
4. Hai đại lượng $X,Y$ độc lập nhau khi và chỉ khi có ràng buộc về xác suất $P_{ij} = P_i P_j$ với mọi $i,j.$

Các khái niệm mới: <br>
1. Covarian: $$\operatorname{cov}(X,Y)=E(XY)-E(X)\cdot E(Y).$$
2. Hệ số tương quan: $$\rho (X,Y)=\frac{\operatorname{cov}(X,Y)}{\sqrt{D(X) \cdot D(Y)}}.$$
Ý nghĩa: đo mức độ phụ thuộc tuyến tính giữa hai đại lượng $X,Y$. Tính chất: $-1\le \rho (X,Y)\le 1.$ <br>
Nếu $X,Y$ độc lập thì $\operatorname{cov}(X,Y)=0$ và kéo theo $\rho (X,Y)=0.$ <br>
Còn nếu $\rho (X,Y)=0$ thì chưa chắc $X,Y$ độc lập. Nếu $\rho (X,Y)\ne 0$ thì chắc chắn $X,Y$ là phụ thuộc.

1. Hãy tính $E(X), E(Y), D(X), D(Y).$ 
* $E(X)$

```python
xy=np.array([[1/6,1/4],[1/6,1/8],[1/6,1/8]])
X=np.array([-1,0,1])
XX=X*X
PX= np.array([xy[0].sum(),xy[1].sum(),xy[2].sum()])
Y= np.array([-1,1])
YY=Y*Y
PY= np.array([xy[:,0].sum(),xy[:,1].sum()])
EX=PX@X
EY= PY@Y
DX=XX@PX -EX**2
DY=YY@PY - EY**2
print('E(X) =',EX)
print('E(Y) =',EY)
print('D(X) =',DX)
print('D(Y) =',DY)
```

```python
covxy = EXY - EX*EY
print('cov(𝑋,𝑌)=',covxy)
```

```python
𝜌xy = covxy/(np.sqrt(DX)*np.sqrt(DY))
print('𝜌(xy) =',𝜌xy)
```
