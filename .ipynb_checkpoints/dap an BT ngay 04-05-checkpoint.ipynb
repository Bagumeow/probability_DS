1. Covarian: $$\operatorname{cov}(X,Y)=E(XY)-E(X)\cdot E(Y).$$
2. Hệ số tương quan: $$\rho (X,Y)=\frac{\operatorname{cov}(X,Y)}{\sqrt{D(X) \cdot D(Y)}}.$$
Ý nghĩa: đo mức độ phụ thuộc tuyến tính giữa hai đại lượng $X,Y$. Tính chất: $-1\le \rho (X,Y)\le 1.$ <br>

```python
#a) Thông thường, biến ngẫu liên rời rạc cho bởi bảng giá trị, tuy nhiên, ở đây cho theo công thức.
#P(X = 0) --> i = 0 thì j = 0 hoặc 1 => P(X=0) = P(X=0,Y=0) + P(X=0,Y=1) = 1/(6.2^0) + 1/(6.2^0) = 1/3
#P(X = 1) --> i = 1 thì j = 0, 1, 2 => tính tổng của 3 số
#P(X = 2) --> i = 2 thì j = 1, 2, 3 => tính tổng của 3 số
#... viết các giá trị ra và dự đoán quy luật
```

```python
#b) P(X = Y | Y < 2) = P(X = Y, Y < 2) / P(Y < 2).
#Tử số = P(X=0,Y=0) + P(X=1, Y=1); 
#Mẫu số = P(Y = 0) + P(Y = 1) = P(X = 0) + P(X = 1) (do vai trò X, Y như nhau) = ... thay số ở câu a)
#c) Các cặp số (X,Y) không âm thỏa mãn điều kiện là: (0,1), (1,0), (1,1), (1,2), (2,1).
#Ta tính tổng của 5 xác suất tương ứng là được.
```

d) Tính P(X=Y) = tổng các xác suất mà X = Y, liệt kê:
* X = Y = 0 -> P(X=0,Y=0) = 1/(6.2^0)
* X = Y = 1 -> P(X=0,Y=0) = 1/(6.2^1)
* X = Y = 2 -> P(X=0,Y=0) = 1/(6.2^2)
* X = Y = 3 -> P(X=0,Y=0) = 1/(6.2^3)
* ...
* tổng quát: X = Y = k -> P(X=k,Y=k) = 1/(6.2^k)
* Khi đó P(X=Y) = 1/6 * (1/2^0 + 1/2^1 + 1/2^2 + ...) trong đó, tổng trong ngoặc là tổng cấp số nhân
* có số hạng đầu u1 = 1, công bội q = 1/2 --> thay vào P = 1/6 * 1/(1/2) = 1/3.

https://www.probabilitycourse.com/chapter5/5_1_6_solved_prob.php
