```python
from scipy import stats as st
```

Bài 1:
Ta có lambda = 250 *0.02

```python
Lambda = 250*0.02
Lambda
```

Gọi X là phân phối poisson với lambda = 5

```python
X = st.poisson(Lambda)
```

```python
print('Xác suất để có đúng 2 máy phế phẩm là',X.pmf(2).round(4))
print('Xác suất để không có không quá 3 phế phẩm là',X.cdf(3).round(4))
```
