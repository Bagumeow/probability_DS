## 2. Các tính chất của các biến cố kết hợp

Trong lab 3, ta đã tìm hiểu một số cách tính mối liên hệ giữa hai biến cố thông qua hệ số tương quan, ma trận hiệp phương sai, ... trong phần này, chúng ta tiếp tục nghiên cứu một số công thức khác về hàm pmf, cdf.
### Vector ngẫu nhiên.
Cho vector ngẫu nhiên $(X,Y)$ gồm một cặp hai biến ngẫu nhiên, ta định nghĩa hàm tích lũy cdf sau đây của phân phối kết hợp $$F(a,b) = P(X \le a, Y \le b).$$
Tính chất: $F(x,y)$ không giảm theo từng biến và $F(-\infty, -\infty)=0, F(+\infty, +\infty) = 1.$ <br>
Khi đó ta sẽ có xác suất lề: $F_X(a) = F(a,+\infty)$ và $F_Y(b) = F(+\infty, b).$ Ta cũng có $$P(a < X \le b, c < Y \le d) = F(a,c) + F(b,d) - F(a,d) - F(b,c).$$


### Trường hợp rời rạc.
Trong trường hợp biến ngẫu nhiên rời rạc, giả sử $E_1,E_2$ lần lượt là không gian mẫu của hai biến ngẫu nhiên $X,Y$, ta sẽ có 
* Hàm mật độ $f(x,y) = P(X=x,Y=y)$ cho bởi bảng giá trị.
* Liên hệ giữa hàm mật độ $f(x,y)$ và hàm phân phối $F(x,y)$ là: $F(x,y) = \sum_{a \le x, b \le y} f(a,b).$
* Xác suất theo từng biến là: $p_X(a) = \sum_{b \in E_2} p(a,b)$ và $p_Y(b) = \sum_{a \in E_1} p(a,b).$
* Xác suất trong một miền: $$P(a_1 < X \le a_2, b_1 < Y \le b_2) = \sum_{a_1 < i \le a_2, b_1 < j \le b_2} p(i,j).$$
* Kỳ vọng có điều kiện: $$E[X|Y=y] = \sum_{x \in E_1} x.p_{X|Y} (x|y)$$ trong đó $p_{X|Y}(x|y) = P(X=x | Y=y) = \frac{p(x,y)}{p_Y(y)}.$

### Trường hợp liên tục.
Trong trường hợp biến ngẫu nhiên liên tục, ta sẽ có
* Hàm mật độ $f(x,y) = P(X=x, Y=y)$ cho bởi hàm số.
* Liên hệ giữa hàm mật độ $f(x,y)$ và hàm phân bố $F(x,y)$ là:
$f(x,y)=\frac{{{d}^{2}}}{dxdy}F(x,y)$ nếu tồn tại đạo hàm và $f(x,y)=0$ trong trường hợp ngược lại. 

* Xác suất riêng theo từng biến (còn gọi là lề) là: $p_X(x) = \int_{-\infty}^{+\infty} f(x,y)dy$ và $p_Y(y) = \int_{-\infty}^{+\infty} f(x,y)dx$ 
* Xác suất trong miền: $$P(a_1 < X < a_2, b_1 < Y < b_2) = \int_{{{a}_{1}}}^{{{a}_{2}}}{\int_{{{b}_{1}}}^{{{b}_{2}}}{f(x,y)dxdy}}.$$
* Kỳ vọng có điều kiện: $$E[X|Y=y]=\int_{-\infty }^{+\infty }{x{{f}_{X|Y}}(x|y)dx}$$ trong đó ${{f}_{X|Y}}(x|y)=\frac{f(x,y)}{{{f}_{Y}}(y)}$. 

### Tính độc lập.
Hai biến ngẫu nhiên $X, Y$ là độc lập khi và chỉ khi $f(x,y) = f_X(x) \cdot f_Y(y)$ với mọi $x,y.$ Như vậy để kiểm tra tính chất này, trước hết ta kiểm tra xác suất lề của từng biến. 
* Trong trường hợp rời rạc, ta sẽ xét một vài giá trị đặc biệt cho trong bảng phân bố xác suất.
* Trường trường hợp liên tục, ta sẽ đối chiếu hai biểu thức của hàm số có được.
